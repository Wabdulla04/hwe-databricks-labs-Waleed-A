## Question 7: Categorize Reviews by Sentiment

Create a query that categorizes reviews into sentiment groups based on their star rating:
- **'Positive'** for ratings of 4 or 5
- **'Neutral'** for rating of 3
- **'Negative'** for ratings of 1 or 2

Show the first 10 reviews with their `star_rating` and a new column called `sentiment`.

Then count how many reviews are in each sentiment category.

```sql
-- TODO: Select star_rating and create a sentiment column
CREATE OR REPLACE TEMPORARY VIEW sentiment_view AS 
SELECT star_rating, 
    CASE WHEN star_rating >= 4 THEN "Positive"
         WHEN star_rating = 3 THEN "Neutral"
         ELSE "Negative" END as sentiment
         FROM reviews;

SELECT * FROM sentiment_view LIMIT 10
```

```sql
-- TODO: Count reviews by sentiment category
SELECT sentiment, COUNT(*) as sentiment_count
FROM sentiment_view
GROUP BY sentiment
```

## Question 8: Most Helpful Review

Find the review with the highest number of helpful votes.

**Hint:** Using the `helpful_votes` column with its current data type will give the wrong answer

```sql
-- TODO: Find the review with the most helpful votes
SELECT product_title, cast(helpful_votes AS INT), cast(total_votes AS INT) FROM reviews 
ORDER BY total_votes DESC
LIMIT 1
```

## Question 9: Date with Most Purchases

Find the purchase date with the most reviews.

Show the date and the count.

```sql
-- TODO: Find the purchase_date with the most reviews
SELECT purchase_date, MAX(cast(total_votes AS INT)) as max_total_votes FROM reviews
GROUP BY purchase_date
ORDER BY max_total_votes DESC
LIMIT 1
```

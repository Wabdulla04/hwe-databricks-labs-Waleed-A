## Exercise 1: Count Employees with Valid Email

Write a query that counts employees with a valid email address.
Valid format: email is not NULL and contains an `@` symbol (e.g., `name@company.com`)
Assign an alias of `valid_email_count` to the count.

```sql
-- @test:valid_email_filter
SELECT COUNT(*) AS valid_email_count FROM week3_testing.employees
WHERE email LIKE "%@%" AND email IS NOT NULL
```

## Exercise 2: Count Employees in Salary Range

Write a query that counts employees with salaries from $50,000 to $100,000 (inclusive).
Assign an alias of `employee_count` to the count.

```sql
-- @test:employees_in_salary_range
SELECT COUNT(*) AS employee_count FROM week3_testing.employees
WHERE salary BETWEEN 50000 AND 100000
```

## Exercise 3: Insert Recent Hires

Insert employees hired after January 1, 2026 into `filtered_employees`.

```sql
-- @test:recent_hires
INSERT INTO week3_testing.filtered_employees
SELECT * FROM week3_testing.employees
WHERE hire_date > DATE("2026-01-01")

```

## Exercise 4: Insert Engineering Employees

Insert only employees from the Engineering department into `filtered_employees`.

```sql
-- @test:engineering_department_filter
INSERT INTO week3_testing.filtered_employees 
SELECT * FROM week3_testing.employees
WHERE department = "Engineering"
```
